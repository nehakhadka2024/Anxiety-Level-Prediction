```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.preprocessing import LabelEncoder
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score,precision_score, recall_score, f1_score, classification_report, confusion_matrix
```

```python
#Loading Dataset
file_path = 'Anxiety.csv' 
data = pd.read_csv(file_path)
```

```python
print("Preview of the first few rows of Dataset")
data.head()
```

```python
# Exploring the data
# Checking the structure and basic details of the dataset

print("\nDataset Overview:")
print(data.info())
print("\nDataset Description:")
print(data.describe())
```

```python
missing_value= data.isnull().sum()
print(missing_value)
print(missing_value[missing_value > 0])
```

```python
plt.figure(figsize=(12, 5))
sns.heatmap(data.isnull(), cbar=False, cmap='coolwarm')
plt.title("Missing Values Heatmap")
plt.show()
```

```python
label_encoder = LabelEncoder()
data['Anxiety Label'] = label_encoder.fit_transform(data['Anxiety Label'])

plt.figure(figsize=(9, 6))
data['Anxiety Label'].value_counts().sort_index().plot(kind='bar',color='skyblue')

# Creating a bar chart showing the count of each anxiety level
plt.title("Distribution of Anxiety Levels")
plt.xlabel("Anxiety Label")
plt.ylabel("Count")

# Displaying the plot
plt.show() 
```

```python
# Boxplot of Numerical Features

numeric_features = data.select_dtypes(include=['int64', 'float64'])

plt.figure(figsize=(12, 6))


# Creating a boxplot 
sns.boxplot(data=numeric_features)

plt.title("Boxplot of Numerical Features")
plt.xticks(rotation=90)

# Displaying the plot
plt.show()
```

```python
# Heatmap of Correlation Matrix
plt.figure(figsize=(8, 5))

# Calculating correlations between numerical features
correlation_matrix = numeric_features.corr()
sns.heatmap(correlation_matrix, annot=True, cmap='viridis', fmt=".2f")
plt.title("Correlation Matrix of Numeric Features")

# Displaying the plot
plt.show()
```

```python
# Density plot of each numerical feature
for column in numeric_features.columns:
 plt.figure(figsize=(8, 5))
    
# Showing the distribution as a smooth curve using KDE
 sns.kdeplot(data=numeric_features[column], fill=True, color='purple')

 plt.title(f"Density Plot of {column}", pad=15)
 plt.xlabel(column, labelpad=10)
plt.ylabel("Density")

# Displaying the plot
plt.show()
    
```

```python
#Boxplots for Anxiety Label and Numerical Features
for column in numeric_features.columns:
    plt.figure(figsize=(8, 5))

    # Boxplot compares how each feature is spread across the different anxiety labels
    sns.boxplot(x=data['Anxiety Label'], y=data[column], hue=data['Anxiety Label'], dodge=False, palette="Set2")
    plt.title(f"Boxplot of {column} by Anxiety Levels")
    plt.xlabel("Anxiety Label")
    plt.ylabel(column)
    plt.legend(title="Anxiety Levels", loc='upper right', labels=label_encoder.classes_)
    plt.xticks(ticks=[0, 1, 2, 3], labels=label_encoder.classes_)

    #Displaying the plot
    plt.show()

```

```python
# Point Plots for Anxiety Label and Numerical Features
for column in numeric_features.columns:
    plt.figure(figsize=(8, 5))
    
    #Showing the average value for each anxiety level
    sns.pointplot(x=data['Anxiety Label'], y=data[column], errorbar='sd')

    plt.title(f"Average {column} by Anxiety Levels")
    plt.xlabel("Anxiety Label")
    plt.ylabel(column)
    plt.xticks(ticks=[0, 1, 2, 3], labels=label_encoder.classes_)

    # Displaying the plot
    plt.show()
```

```python
# Bar Plot of Average Feature Values by Anxiety Label

#Calculating mean of each numerical feature of each anxiety label
avg_features = numeric_features.groupby(data['Anxiety Label']).mean()
avg_features.T.plot(kind='bar', figsize=(14, 7), colormap="plasma")
plt.title("Mean Feature Values by Anxiety Level")
plt.xlabel("Features")
plt.ylabel("Mean Value")
plt.xticks(rotation=45)
plt.show()
```

```python
#Train-Test Split

#Selecting the input features
features = data.iloc[:, 7:14]
target = data['Anxiety Label']
X_train, X_test, y_train, y_test = train_test_split(features, target, test_size=0.2, random_state=42, stratify=target
)
```

```python
#Initializing the Models
models = {
    "Logistic Regression": LogisticRegression(max_iter=1000, random_state=42),
    "Decision Tree": DecisionTreeClassifier(random_state=42),
    "Random Forest": RandomForestClassifier(random_state=42)
}
```

```python
# Models and evaluation training
for model_name, model in models.items():
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)

    # Calculating evaluation metrics
    accuracy = accuracy_score(y_test, y_pred)

    # Changing the 'weighted' for multiclass
    precision = precision_score(y_test, y_pred, average='weighted') 
    recall = recall_score(y_test, y_pred, average='weighted') 
    f1 = f1_score(y_test, y_pred, average='weighted') 

    # Displaying the result/metrics
    print(f"\nMetrics for {model_name}:")
    print(f"Accuracy: {accuracy:.2f}")
    print(f"Precision (Weighted): {precision:.2f}") 
    print(f"Recall (Weighted): {recall:.2f}") 
    print(f"F1-Score (Weighted): {f1:.2f}") 
```

```python
# Train, Predict, and Evaluate
results = []

for model_name, model in models.items():
    
    # Training the model using the training data
    model.fit(X_train, y_train)
    
    # Using the trained model to predict the labels
    y_pred = model.predict(X_test)
    
    # Calculating the accuracy
    accuracy = accuracy_score(y_test, y_pred)
    
    # Creating a detailed classification report
    # Converting class names to strings avoids errors if labels are numeric
    target_names = [str(label) for label in label_encoder.classes_]
    report = classification_report(y_test, y_pred, target_names=target_names)
    
    # Saving results
    results.append({"Model": model_name, "Accuracy": accuracy, "Report": report})

    # Confusion Matrix
    plt.figure(figsize=(10, 7))
    cm = confusion_matrix(y_test, y_pred)
    sns.heatmap(cm, annot=True, fmt="d", cmap="viridis", xticklabels=label_encoder.classes_, yticklabels=label_encoder.classes_)
    plt.title(f"Confusion Matrix for: {model_name}")
    plt.xlabel("Predicted label")
    plt.ylabel("Actual label")

    #Displaying the plot
    plt.show()
```

```python
#Cross-Validation
cv_results = {}
print("\nCross-Validation Results:")

for model_name, model in models.items():
    scores = cross_val_score(model, features, target, cv=5, scoring='accuracy')
    cv_results[model_name] = scores
    print(f"{model_name}: Mean CV Accuracy = {scores.mean():.4f}, Scores = {scores}")


```

```python

#Average Cross-Validation Accuracy (Bar Chart)
cv_df = pd.DataFrame(cv_results)

plt.figure(figsize=(10, 6))
cv_df.mean().plot(kind='bar', yerr=cv_df.std(),colormap="plasma")

plt.title("Cross-Validation accuracy scores for each Models")
plt.xlabel("Model")
plt.ylabel("Accuracy Score")
plt.xticks(rotation=45)

#Displaying the plot
plt.show()
```

```python
# Top Features Influencing Tree-Based Models
for model_name, model in models.items():
    if hasattr(model, "feature_importances_"):
        plt.figure(figsize=(6, 4))

        # Converting the importances to pandas Series to sort and plot them easily
        importance = pd.Series(model.feature_importances_, index=features.columns)

        # Plotting the most important features first
        importance.sort_values(ascending=False).plot(kind='bar', color='teal')
        plt.title(f"Top Features Influencing {model_name}")
        plt.ylabel("Importance")
        plt.xlabel("Feature Names")

        #Displaying the plot
        plt.show()
```

```python
# Results summarized
results_df = pd.DataFrame(results, columns=["Model", "Accuracy", "Report"])
print("\nSummary of Model Performance:")
print(results_df[["Model", "Accuracy"]])
```

```python
import numpy as np
import pandas as pd

# Same feature columns the model was trained on
feature_columns = features.columns

# Function to take user input and predict anxiety level
def predict_anxiety_level(model, feature_columns):
    print("\n--- Anxiety Level Prediction ---")
    user_input = []

    # Ask user for each feature value
    for feature in feature_columns:
        while True:
            try:
                value = float(input(f"Enter value for '{feature}': "))
                user_input.append(value)
                break
            except ValueError:
                print("Invalid input. Please enter a numeric value.")

    # Convert input to DataFrame (important for ML models)
    input_df = pd.DataFrame([user_input], columns=feature_columns)

    # Predict anxiety level
    prediction = model.predict(input_df)
    predicted_label = label_encoder.inverse_transform(prediction)

    print("\n--- Prediction Result ---")
    print(f"Predicted Anxiety Level: {predicted_label[0]}")

# Run prediction
predict_anxiety_level(models["Random Forest"], feature_columns)
```
